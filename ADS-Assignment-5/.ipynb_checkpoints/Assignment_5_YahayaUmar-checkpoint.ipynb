```python
import pandas as pd
import plotly.express as px

import warnings
warnings.filterwarnings('ignore')
```

```python
data=pd.read_csv('titanic data.csv')
```

```python
data.shape
```

```python
data.head()
```

```python
data.isnull().sum()
```

```python
# Handling Null Values

data['Age'] = data['Age'].fillna(0)
data['Cabin'] = data['Cabin'].fillna("Not Specified")
data['Embarked'] = data['Embarked'].fillna('Not Specified')
```

```python
data.isnull().sum()
```

```python
data.info()
```

```python
q1 = px.histogram(data, x="Age", nbins=20, 
                title='Histogram of Age',
                opacity=0.8,
                color_discrete_sequence=['indianred'])
q1.show()
```

```python
fig = px.histogram(data, x="Age", color='Sex', nbins=20)
fig.show()
```

```python
 data.corr()
```

```python
px.scatter_matrix(data)
```

```python
#pivot table 

pivot1=pd.pivot_table(data,values='Fare',columns='Pclass', margins=True)
pivot1
```
